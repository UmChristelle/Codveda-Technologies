# Task 1: Data Cleaning

This notebook loads the Iris dataset, checks for nulls and duplicates, cleans the data, and saves the cleaned file.

```python
from pathlib import Path

import pandas as pd
```

```python
base_dir = Path.cwd()
if not (base_dir / "data" / "iris.csv").exists():
    base_dir = Path("Task_1_Data_Cleaning")

data_path = base_dir / "data" / "iris.csv"
output_dir = base_dir / "outputs"
output_dir.mkdir(parents=True, exist_ok=True)

data_path
```

```python
df = pd.read_csv(data_path)
print("Original shape:", df.shape)
print("\nColumn data types:")
print(df.dtypes)
df.head()
```

## Check Nulls and Duplicates

```python
print("Null values per column:")
print(df.isnull().sum())
print("\nTotal null values:", int(df.isnull().sum().sum()))
print("Duplicate rows:", int(df.duplicated().sum()))
```

## Clean the Data

```python
df_clean = df.drop_duplicates().copy()

numeric_columns = df_clean.select_dtypes(include="number").columns
for column in numeric_columns:
    df_clean[column] = df_clean[column].fillna(df_clean[column].mean())

categorical_columns = df_clean.select_dtypes(exclude="number").columns
for column in categorical_columns:
    df_clean[column] = df_clean[column].astype(str).str.strip().str.lower()
    mode = df_clean[column].mode()
    if not mode.empty:
        df_clean[column] = df_clean[column].fillna(mode.iloc[0])

print("Cleaned shape:", df_clean.shape)
print("\nNull values after cleaning:")
print(df_clean.isnull().sum())
print("\nUnique categorical values after standardization:")
for column in categorical_columns:
    print(f"{column}: {sorted(df_clean[column].unique().tolist())}")
```

## Save Cleaned File

```python
output_file = output_dir / "iris_cleaned.csv"
df_clean.to_csv(output_file, index=False)
print("Saved to:", output_file)
df_clean.head()
```
